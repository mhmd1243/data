### Dataset Description

This dataset contains COVID-19-related information with the following key attributes:

- **_id**: A unique identifier for each record (14,911 unique values).
- **Outbreak Associated**: Indicates whether the case is associated with an outbreak (2 unique values).
- **Age Group**: The age group of the individuals (9 unique values).
- **Neighbourhood Name**: The name of the neighborhood (140 unique values).
- **FSA (Forward Sortation Area)**: The first three characters of the postal code, representing geographic areas (96 unique values).
- **Source of Infection**: The origin of the infection (8 unique values).
- **Classification**: Classification of the case (2 unique values).
- **Episode Date**: The date of the episode (149 unique values).
- **Reported Date**: The date the case was reported (142 unique values).
- **Client Gender**: Gender of the individual (5 unique values).
- **Outcome**: Outcome of the case, e.g., recovered, deceased, etc. (3 unique values).
- **Currently Hospitalized**: Indicates if the individual is currently hospitalized (2 unique values).
- **Currently in ICU**: Indicates if the individual is currently in the ICU (2 unique values).
- **Currently Intubated**: Indicates if the individual is currently intubated (2 unique values).
- **Ever Hospitalized**: Indicates if the individual has ever been hospitalized (2 unique values).
- **Ever in ICU**: Indicates if the individual has ever been in the ICU (2 unique values).
- **Ever Intubated**: Indicates if the individual has ever been intubated (2 unique values).

This dataset provides comprehensive information on COVID-19 cases, including demographic details, hospitalization status, and geographic information, which can be used for analysis and visualization.

```python
!pip install geopy

import pandas as pd
from geopy.geocoders import Nominatim
import time

# List of Toronto neighborhoods
neighborhoods = [
    'Malvern', 'Rouge', 'Centennial Scarborough', 'Highland Creek', 'West Hill', 'Guildwood', 'Morningside', 
    'Woburn', 'Scarborough Village', 'Eglinton East', 'Bendale', 'Clairlea-Birchmount', 'Kennedy Park', 'Ionview', 
    'Oakridge', 'Wexford/Maryvale', 'Cliffcrest', 'Birchcliffe-Cliffside', 'Dorset Park', 'Victoria Village', 
    'Agincourt South-Malvern West', "L'Amoreaux", 'Agincourt North', "Tam O'Shanter-Sullivan", 'Milliken', 'Steeles', 
    'Bayview Woods-Steeles', 'Hillcrest Village', 'Pleasant View', 'Henry Farm', 'Don Valley Village', 'Bayview Village', 
    'St.Andrew-Windfields', 'Bridle Path-Sunnybrook-York Mills', 'Newtonbrook West', 'Newtonbrook East', 'Willowdale East', 
    'Willowdale West', 'Lansing-Westgate', 'Bay Street Corridor', 'Westminster-Branson', 'Parkwoods-Donalda', 
    'Banbury-Don Mills', 'Flemingdon Park', "O'Connor-Parkview", 'Clanton Park', 'Bathurst Manor', 
    'York University Heights', 'Downsview-Roding-CFB', 'Glenfield-Jane Heights', 'Black Creek', 'Taylor-Massey', 
    'East End-Danforth', 'Woodbine Corridor', 'Danforth-East York', 'Woodbine-Lumsden', 'Old East York', 'Danforth', 
    'The Beaches', 'Leaside-Bennington', 'Thorncliffe Park', 'Blake-Jones', 'Greenwood-Coxwell', 'North Riverdale', 
    'Playter Estates-Danforth', 'Broadview North', 'South Riverdale', 'Lawrence Park South', 'Lawrence Park North', 
    'Bedford Park-Nortown', 'Mount Pleasant West', 'Mount Pleasant East', 'Yonge-Eglinton', 'Rosedale-Moore Park', 
    'Casa Loma', 'Yonge-St.Clair', 'Forest Hill South', 'Church-Yonge Corridor', 'North St. James Town', 
    'Cabbagetown-South St. James Town', 'Moss Park', 'Regent Park', 'Waterfront Communities-The Island', 'Forest Hill North', 
    'Humewood-Cedarvale', 'Annex', 'Wychwood', 'University', 'Kensington-Chinatown', 'Niagara', 'Yorkdale-Glen Park', 
    'Englemount-Lawrence', 'Caledonia-Fairbank', 'Briar Hill - Belgravia', 'Beechborough-Greenbrook', 'Oakwood Village', 
    'Corso Italia-Davenport', 'Palmerston-Little Italy', 'Dovercourt-Wallace Emerson-Junction', 'Dufferin Grove', 
    'Little Portugal', 'Trinity-Bellwoods', 'South Parkdale', 'Roncesvalles', 'Maple Leaf', 'Brookhaven-Amesbury', 
    'Rustic', 'Mount Dennis', 'Keelesdale-Eglinton West', 'Weston-Pellam Park', 'Rockcliffe-Smythe', 'Junction Area', 
    'High Park North', 'High Park-Swansea', 'Runnymede-Bloor West Village', 'Lambton Baby Point', 'Mimico (includes Humber Bay Shores)', 
    'New Toronto', 'Long Branch', 'Alderwood', 'Stonegate-Queensway', 'Kingsway South', 'Islington-City Centre West', 
    'Edenbridge-Humber Valley', 'Princess-Rosethorn', 'Eringate-Centennial-West Deane', 'Etobicoke West Mall', 'Markland Wood', 
    'Humber Summit', 'Humbermede', 'Pelmo Park-Humberlea', 'Weston', 'Humber Heights-Westmount', 
    'Willowridge-Martingrove-Richview', 'Kingsview Village-The Westway', 'Thistletown-Beaumond Heights', 
    'Mount Olive-Silverstone-Jamestown', 'West Humber-Clairville', 'Rexdale-Kipling', 'Elms-Old Rexdale'
]

# Initialize geolocator
geolocator = Nominatim(user_agent="toronto_neighborhood_geocoder")

# Prepare lists for storing results
neighborhood_names = []
latitudes = []
longitudes = []

# Loop over the neighborhoods and fetch their latitude and longitude
for neighborhood in neighborhoods:
    try:
        location = geolocator.geocode(f"{neighborhood}, Toronto, Canada")
        if location:
            neighborhood_names.append(neighborhood)
            latitudes.append(location.latitude)
            longitudes.append(location.longitude)
        else:
            neighborhood_names.append(neighborhood)
            latitudes.append(None)
            longitudes.append(None)
    except Exception as e:
        print(f"Error with {neighborhood}: {e}")
        neighborhood_names.append(neighborhood)
        latitudes.append(None)
        longitudes.append(None)
    
    # Sleep to avoid hitting the API too quickly
    time.sleep(1)

# Create a DataFrame
df = pd.DataFrame({
    "Neighborhood": neighborhood_names,
    "Latitude": latitudes,
    "Longitude": longitudes
})

# Save DataFrame to CSV
csv_file = 'toronto_neighborhood_latitudes_longitudes.csv'
df.to_csv(csv_file, index=False)

csv_file
```
