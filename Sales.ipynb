### Dataset Columns:
#### explanation file

1. **InvoiceNo**: A unique identifier for each invoice or order.
2. **StockCode**: A unique code associated with the item being purchased.
3. **Description**: A description of the product or item.
4. **Quantity**: The number of units purchased in the order.
5. **InvoiceDate**: The date and time when the order was placed.
6. **UnitPrice**: The price per unit of the item purchased.
7. **CustomerID**: The unique identifier for the customer making the purchase.
8. **Country**: The country of the customer.
9. **Discount**: The discount applied to the order, if any.
10. **PaymentMethod**: The method used to pay for the order (e.g., Credit Card, Bank Transfer).
11. **ShippingCost**: The cost of shipping the order, if available.
12. **Category**: The category of the product (e.g., Furniture, Apparel, Stationery).
13. **SalesChannel**: Indicates whether the order was placed online or in-store.
14. **ReturnStatus**: The return status of the order (e.g., Not Returned).
15. **ShipmentProvider**: The shipping provider used for the delivery of the order (e.g., UPS, FedEx).
16. **WarehouseLocation**: The location of the warehouse that fulfilled the order.
17. **OrderPriority**: The priority level of the order (e.g., High, Medium, Low).


```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
#handle date in first stage
pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
df=pd.read_csv("online_sales_dataset.csv")
```

```python
df.sample(10)
```

#### The data exhibits several mistakes at first look, beginning with a negative quantity.

```python
df=df.query("`Quantity`>0")
```

```python
df.info()
```

#### Second, the datetime type is handled by the date columns.

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
#cheaking For duplicates
df.duplicated().sum()
```

#### no duplicates were found

```python
#cheak for nulls
df.isna().sum()
```

#### I need to view the data of the null values to decide how to treat them.

```python
df[df['CustomerID'].isna()]
```

#### It appears that the null values are ineffective so that other columns are correct,
#### thus I will manage them without removing with the following topolgy,
#### swapping CustomerID with 0 and WarehouseLocation with UNKNOWN.

```python
df['CustomerID'].fillna(0,inplace=True)
df['WarehouseLocation'].fillna('UNKNOWN',inplace=True)
```

```python
df.isna().sum()
```

#### Time extract columns (day, month, quarter, and year) for future use in plotting and distribution analysis.

```python
df['Year']=df['InvoiceDate'].dt.year
df['Quarter']=df['InvoiceDate'].dt.quarter
df['Month']=df['InvoiceDate'].dt.month_name()
df['Day']=df['InvoiceDate'].dt.day_name()
```

```python
df.sample(5)
```

#### Brief statistical description for numeric columns

```python
df.columns
```

```python
df[['Quantity','UnitPrice','Discount','ShippingCost']].describe()
```

```python
df['Sales'] = round(((df['UnitPrice'] * df['Quantity']) - (df['ShippingCost'] + (df['UnitPrice'] * df['Quantity'] * df['Discount'])).astype(int))/1000,2)


pivot_table = pd.pivot_table(df, values='Sales', index=['Year', 'Quarter', 'Month'], aggfunc='sum')


display(pivot_table)
```
